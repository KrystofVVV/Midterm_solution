```python
# Create an empty dictionary to store the highest and lowest prices for each company
prices = {}

# Iterate through all CSV files in the "dataset" folder
for file in os.listdir("data"):
    if file.endswith(".csv"):
        file_path = os.path.join("data", file)
        ticker = file[:-4]  # Remove the ".csv" extension to get the ticker

        # Read the CSV file, parse the 'Date' column as a datetime variable
        data = pd.read_csv(file_path, parse_dates=["Date"])

        # Get the highest and lowest Adj Close prices for the current ticker
        highest_price = data["Adj Close"].max()
        lowest_price = data["Adj Close"].min()

        # Add the highest and lowest prices to the dictionary
        prices[ticker] = {"Highest Adj Close": highest_price, "Lowest Adj Close": lowest_price}

# Print the highest and lowest Adj Close prices for each company
for ticker in prices:
    print(f"Ticker: {ticker}")
    print(f"Highest Adj Close: {prices[ticker]['Highest Adj Close']}")
    print(f"Lowest Adj Close: {prices[ticker]['Lowest Adj Close']}\n")
```

```python
import numpy as np

returns = {}

# Iterate through all CSV files in the "dataset" folder
for file in os.listdir("data"):
    if file.endswith(".csv"):
        file_path = os.path.join("data", file)
        ticker = file[:-4]  # Remove the ".csv" extension to get the ticker

        # Read the CSV file, parse the 'Date' column as a datetime variable
        data = pd.read_csv(file_path, parse_dates=["Date"])

        # Calculate logarithmic returns from Adj Close
        data["Log_Ret"] = np.log(data["Adj Close"]).diff()

        # Report on min, max, mean, and median of the return distribution
        min_return = data["Log_Ret"].min()
        max_return = data["Log_Ret"].max()
        mean_return = data["Log_Ret"].mean()
        median_return = data["Log_Ret"].median()

        returns[ticker] = {"min": min_return, "max": max_return, "mean": mean_return, "median": median_return}

# Print the results for each company
for ticker, stats in returns.items():
    print(f"Ticker: {ticker}")
    print(f"Min Log Return: {stats['min']:.4f}")
    print(f"Max Log Return: {stats['max']:.4f}")
    print(f"Mean Log Return: {stats['mean']:.4f}")
    print(f"Median Log Return: {stats['median']:.4f}")
    print("\n")
```

```python
highest_gain = {}
highest_loss = {}

# Iterate through all CSV files in the "dataset" folder
for file in os.listdir("data"):
    if file.endswith(".csv"):
        file_path = os.path.join("data", file)
        ticker = file[:-4]  # Remove the ".csv" extension to get the ticker

        # Read the CSV file, parse the 'Date' column as a datetime variable
        data = pd.read_csv(file_path, parse_dates=["Date"])

        # Calculate logarithmic returns from Adj Close
        data["Log_Ret"] = np.log(data["Adj Close"]).diff()

        # Find the date with the highest gain and loss
        highest_gain_date = data.loc[data["Log_Ret"].idxmax(), "Date"].strftime("%Y-%m-%d")
        highest_loss_date = data.loc[data["Log_Ret"].idxmin(), "Date"].strftime("%Y-%m-%d")

        highest_gain[ticker] = highest_gain_date
        highest_loss[ticker] = highest_loss_date
tickers = ['BKNG','COF','EVRG','SYK']
# Print the results for each company
for ticker in tickers:
    print(f"Ticker: {ticker}")
    print(f"Highest Gain Date: {highest_gain[ticker]}")
    print(f"Highest Loss Date: {highest_loss[ticker]}")
    print("\n")
```

```python
tickers = ['BKNG','COF','EVRG','SYK']

weekly_volumes = {}

for ticker in tickers:
    # Read the CSV file, parse the 'Date' column as a datetime variable
    data = pd.read_csv(f"data/{ticker}.csv", parse_dates=["Date"])

    # Resample the data by week and calculate the mean of the weekly volume
    weekly_data = data.resample('W', on='Date').mean()
    weekly_volumes[ticker] = weekly_data['Volume'].mean()

# Print the average weekly volume for each company
for ticker, volume in weekly_volumes.items():
    print(f"Ticker: {ticker}")
    print(f"Average weekly volume: {volume:.2f}")
    print("\n")
```

```python
tickers = ['BKNG', 'COF', 'EVRG', 'SYK']

total_returns = {}

for ticker in tickers:
    # Read the CSV file, parse the 'Date' column as a datetime variable
    data = pd.read_csv(f"data/{ticker}.csv", parse_dates=["Date"])

    # Calculate the logarithmic return from Adj Close
    data["Log_Ret"] = np.log(data["Adj Close"]).diff()

    # Calculate the total return
    total_return = (np.exp(data["Log_Ret"].sum()) - 1) * 100

    total_returns[ticker] = total_return

# Find the company with the highest total return
best_company = max(total_returns, key=total_returns.get)

print(f"The company with the highest total return is {best_company} with a total return of {total_returns[best_company]:.2f}%")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))

for ticker in tickers:
    # Read the CSV file, parse the 'Date' column as a datetime variable
    data = pd.read_csv(f"data/{ticker}.csv", parse_dates=["Date"])

    # Calculate logarithmic returns from Adj Close
    data["Log_Ret"] = np.log(data["Adj Close"]).diff()

    # Plot the log-returns
    data.plot(x='Date', y='Log_Ret', label=ticker, ax=ax)

ax.set(title='Log-returns of companies', xlabel='Date', ylabel='Log-returns')
ax.legend(loc='upper left')

plt.show()
```

```python


# Read the CSV file from URL
url = "https://ies-python-midterm.s3.eu-central-1.amazonaws.com/companies/companies_no_subindustry.csv"
df = pd.read_csv(url)

# Count the number of companies with missing inclusion date
num_missing_dates = df["included"].isna().sum()
print(f"Number of companies with missing inclusion date: {num_missing_dates}")

# Drop rows with missing inclusion date and convert "included" column to datetime
df = df.dropna(subset=["included"])
df["included"] = pd.to_datetime(df["included"], dayfirst=True, errors="coerce")

# Calculate the oldest and youngest constituent
oldest_constituent = df.loc[df["included"].idxmin()]
youngest_constituent = df.loc[df["included"].idxmax()]

print(f"Oldest constituent: {oldest_constituent['name']} ({oldest_constituent['included'].strftime('%Y-%m-%d')})")
print(f"Youngest constituent: {youngest_constituent['name']} ({youngest_constituent['included'].strftime('%Y-%m-%d')})")

# Calculate the average age of a constituent in the sample
now = pd.Timestamp.now()
df["age"] = (now - df["included"]).dt.days / 365
avg_age = df["age"].mean()
print(f"Average age of a constituent: {avg_age:.2f} years")

#I dont know why this code doesnt work
```
